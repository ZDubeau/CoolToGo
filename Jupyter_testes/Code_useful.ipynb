```python
À propos du projet " Cool To Go ! "
Génerale
Buts

Le but premier de l'application est d'aider les gens à découvrir des activités, dans ce cas, 
ce sont des activités qui peuvent aider les gens à se rafraîchir pendant les périodes de forte 
chaleur et de canicule.

Le climat grenoblois s’est réchauffé sur ces 50 dernières années et la ville subit chaque été 
des vagues de chaleurs plus intenses et plus fréquentes. La tendance va certainement se renforcer 
avec une augmentation attendue des épisodes caniculaires

Les études climatiques grenobloises prévoient des températures de plus en plus hautes en saison estivale. 
Le tout accompagné d’un passage de 3 jours à 46 jours de forte chaleur d’ici 30 ans !

Le projet a été conçu avec l'idée de profil, mode de transport et temps disponible. 
La localisation de l'utilisateur est également relative, car l'idée est de permettre 
à quelqu'un de découvrir une activité à laquelle il peut participer pour échapper à la chaleur : 
"Je surchauffe, j'ai besoin de me rafraîchir... quelles sont les options pour quelqu'un comme moi ?"

Le profil était un point de données important, car toutes les activités ne conviennent pas 
à tous les types de personnes - une aire d'exercice en plein air peut être intéressante pour 
les adultes, mais les séniors peuvent ne pas trouver cette activité intéressante, ou même sécuritaire.
Technique

La détermination des activités auxquelles une personne peut participer dépend de l'endroit 
où elle se trouve, de son mode de transport et du temps dont elle dispose ; lorsqu'un utilisateur 
consulte les activités disponibles, il voit l'ensemble des activités filtrées par son type de profil, 
ainsi que que par l'accès par son mode de transport.

Pour déterminer ce filtre basé sur l'accessibilité des lieux, une API fournie par Targomo a été 
utilisée pour mesurer des temps de parcours précis vers tous les lieux, dont les résultats sont 
utilisés pour filtrer, mais aussi pour montrer le temps nécessaire pour atteindre l'activité dans 
l'application, ainsi que pour montrer le trajet proposé sur la carte.
Contraintes

Dans cette application, il y avait deux contraintes principales : le temps et les données.

Le projet a été réalisé par un seul développeur, dans son temps libre, et de nombreux choix 
ont été faits en fonction des contraintes de temps :

Tous les attributs des lieux ne sont pas affichés - il s'agit de plus d'informations encore cachées !
En version "desktop", la vue liste et la vue carte pourraient très bien être sur la même page, 
et ce serait mieux. Il aurait fallu beaucoup plus de temps pour l'assembler de cette façon.
Les routes pourraient être décomposées en segments, et affichées comme une véritable application 
de recherche de route - avec tranferts et numéro de bus, etc.

De plus, les données ont été assemblées manuellement, donc il y a beaucoup de limitations :

incohérences dans la taille, la résolution et la qualité de l'image
les limites de la portée géographique des données - il y a beaucoup de choses qui sont atteignables 
depuis Grenoble qui ne sont pas incluses dans le jeu de données (les activités de montagne, par exemple).
il aurait été extrêmement intéressant de pouvoir évaluer les activités en fonction de leur niveau 
de "fraîcheur".
les lieux commerciaux sont également intéressants pour échapper à une canicule, et le fait 
d'avoir des données privées/commerciales supplémentaires rendrait l'application plus complète et utile.

Attribution
Temps de transport


Icones

elderly, father and children, jump, people By Adrien Coquet, FR In the Peoples Collection
hands and heart By Eliricon, FI In the hand and palm Collection
```
